#### Set Working Directory & Accessing Data 

```python
# set working directory
os.chdir('/Users/ElleryBresler/Documents/GitHub/week8labs/data')  # add your filepath within the quotes
gh_turnover_four_projects = pd.read_csv("gh_turnover_four_projects.csv") # load the file
```

#### Rename Columns

```python
gh_turnover_four_projects = (gh_turnover_four_projects
 .rename(columns={'domain':'project_domain',
                  'language':'project_language',
                  'windows':'project_age',
                  'window_idx':'quarter',
                  'num_team':'team_size',
                  'num_commits':'project_commits',
                  'blau_gender':'gender_blau',
                  'Gini_gh_ten':'gh_tenure_gini',
                  'Core1':'core_dev',
                  'commits':'user_commits',
                  'propCommits':'commit_proportion',
                  'github_tenure':'gh_tenure',
                  'leavesNextQ':'leaves_next_q'})
       )
```

#### Data Wrangling

```python
#finding the mean and median of user_commits
(gh_turnover_four_projects
.agg({
    'user_commits':['mean', 'median']})
)
```

```python
#finding columns that start with p or contain a g
turnover_new = (gh_turnover_four_projects
                .filter(regex = "(^p|g)")
               )
```

```python
#finding rows that have project_commits greater than 9000
(turnover_new.query("project_commits > 9000"))
```

```python
#using piping notation to do it all at once 
turnover_newer = (gh_turnover_four_projects 
                  .filter(regex = "(^p|g)")
                  .query("project_commits > 9000")
                 )

turnover_newer
```

```python
#using piping notation to select columns that end with e or contain s 
#selectig rows the gh_tenure is greater than or equal to 1991
(gh_turnover_four_projects
 .filter(regex = "(e$|s)")
 .query("gh_tenure >= 1991")
)
```

```python
(gh_turnover_four_projects
 .filter(regex = "(e$|s)")
 .query("gh_tenure >= 1991")
 .agg({'user_commits':['mean', 'median']})
)
```
